```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import csv
```

```python
# Load in csv

file=os.path.join("avg_rain_state.csv")
avg_rain=pd.read_csv(file)
avg_rain.head()
```

```python
# Set x axis and tick locations

x_axis = np.arange(len(avg_rain))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space

plt.figure(figsize=(25,10))
plt.bar(x_axis, avg_rain["Inches"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, avg_rain["State"], rotation="vertical")
```

```python
# Set x and y limits

plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(avg_rain["Inches"])+10)
```

```python
# Set a Title and labels

plt.title("Average Rain per State")
plt.xlabel("State")
plt.ylabel("Average Amount of Rainfall in Inches")
```

```python
# Save our graph and show the graph

plt.savefig("Average Rain Per State.pdf")
plt.show()
```
